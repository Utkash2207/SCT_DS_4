```python
df1=df
```

```python
df1['Accident_Index'] = df1['Accident_Index'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
```

```python
df1.drop(['Vehicle_Type','Urban_or_Rural_Area','Weather_Conditions'],axis=1,inplace=True)
```

```python
df_num = df1.select_dtypes(include=np.number)
plt.figure(figsize=(15 ,9))
sns.heatmap(df_num.corr() , annot=True)


```

```python
fig, ax = plt.subplots(figsize=(8,5))
Number_of_Vehicles = df1['Local_Authority_(District)'].value_counts()
Number_of_Vehicles[:10].plot(kind='bar')
ax.set(title = 'Number of Accidents',
       xlabel = 'Local_Authority_(District)',
       ylabel = 'Accidents_Index')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
label = [1,2,3,4]

accidents_severity = df1['Accident_Severity'].value_counts()
plt.pie(accidents_severity, labels=label,
        autopct='%1.1f%%', pctdistance=0.85)
circle = plt.Circle( (0,0), 0.5, color='white')
p=plt.gcf()
p.gca().add_artist(circle)
ax.set_title("Accident by Severity",fontdict={'fontsize': 16})
plt.tight_layout()
plt.show()
```

```python
df1['Time'] = pd.to_datetime(df1['Time'])

# Now you can access the hour using dt.hour
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(df1['Time'].dt.hour, bins=24)

plt.xlabel("Accident_Index")
plt.ylabel("Number_of_Casualties")
plt.title("Accidents Count By Time of Day")

plt.show()
```

```python

df=pd.read_csv('/content/Road Accident Data.csv')
weather_conditions = df['Weather_Conditions'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
weather_conditions.sort_values(ascending=False)[:20].plot(kind='bar')
ax.set(title='Weather Conditions at Time of Accident Occurence',
       xlabel='Weather',
       ylabel='Accidents Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Handle missing values and ensure numeric data type

# 1. Drop rows with missing values in 'Accident_Severity' and 'Latitude'
# Assuming 'Accident_Severity' and 'Latitude' are the intended columns
# The original df1 DataFrame likely contains 'Accident_Severity'
df_num = df1.dropna(subset=['Accident_Severity', 'Latitude'])  # Changed to df1

# 2. Convert 'Accident_Severity' and 'Latitude' to numeric if needed
df_num['Accident_Severity'] = pd.to_numeric(df_num['Accident_Severity'], errors='coerce')
df_num['Latitude'] = pd.to_numeric(df_num['Latitude'], errors='coerce')

# Drop any rows that still have missing values after the conversion
df_num = df_num.dropna(subset=['Accident_Severity', 'Latitude'])

# Now create the scatter plot
df_num.plot(kind='scatter', y='Latitude', x='Accident_Severity') # Changed to 'Latitude'
plt.title('Scatter Plot of Start_Lat vs Severity')
plt.xlabel('Severity')
plt.ylabel('Start_Lat')
plt.show()
```
